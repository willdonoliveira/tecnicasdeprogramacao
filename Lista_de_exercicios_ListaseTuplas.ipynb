```python
#TAREFA 4
#Faça um programa que solicita uma quantidade indefinida de notas de uma prova (deve parar quando um número negativo for digitado). Armazena em uma lista. Após isso, o programa deve mostrar a média das notas.
notas = []

while True:
    nota = float(input("Digite uma nota: "))

    if nota < 0:
        break

    notas.append(nota)

if len(notas) > 0:
    media = sum(notas) / len(notas)
    print(f"\nNotas digitadas: {notas}")
    print(f"Média das notas: {media:.2f}")
else:
    print("Nenhuma nota válida foi digitada.")
```

```python
#TAREFA 5
#Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for digitado). Armazene-os em duas listas: uma para os números pares; outra para os números ímpares. A seguir, mostre a porcentagem de pares e ímpares digitados.
pares = []
impares = []

while True:
    num = int(input("Digite um número: "))

    if num < 0:
        break

    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)

total = len(pares) + len(impares)

if total > 0:
    perc_pares = (len(pares) / total) * 100
    perc_impares = (len(impares) / total) * 100

    print(f"\nNúmeros pares: {pares}")
    print(f"Números ímpares: {impares}")

    print(f"\nPorcentagem de pares: {perc_pares:.2f}%")
    print(f"Porcentagem de ímpares: {perc_impares:.2f}%")
else:
    print("Nenhum número positivo foi digitado.")
```

```python
#TAREFA 6
#Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for digitado). Armazene todos os números digitados em uma lista, sem repetição.
numeros = []

while True:
    num = int(input("Digite um número: "))

    if num < 0:
        break

    if num not in numeros:
        numeros.append(num)

print("\nNúmeros digitados (sem repetição):")
print(numeros)
```

```python
#TAREFA 7
#Escreva um programa que solicita que o usuário digite a altura e o peso de 5 pessoas. Verifique se pelo menos duas das pessoas tem a mesma altura e mesmo peso.
pessoas = []

for i in range(5):
    altura = float(input(f"Digite a altura da pessoa {i+1}: "))
    peso = float(input(f"Digite o peso da pessoa {i+1}: "))

    pessoas.append((altura, peso))

duplicado = False
for i in range(len(pessoas)):
    for j in range(i + 1, len(pessoas)):
        if pessoas[i] == pessoas[j]:
            duplicado = True
            break

if duplicado:
    print("\nPelo menos duas pessoas têm a mesma altura e o mesmo peso.")
else:
    print("\nNenhuma dupla com altura e peso iguais.")
```

```python
#TAREFA 8
#Faça um programa que solicita que o usuário digite o nome e a idade de diversas pessoas (o programa deve parar quando um nome vazio for informado). Retorne o nome da pessoa mais velha.
mais_velho_nome = ""
mais_velho_idade = -1

while True:
    nome = input("Digite o nome: ")

    if nome == "":
        break  # encerra o programa

    idade = int(input("Digite a idade: "))

    if idade > mais_velho_idade:
        mais_velho_idade = idade
        mais_velho_nome = nome

if mais_velho_nome != "":
    print(f"\nA pessoa mais velha é: {mais_velho_nome} ({mais_velho_idade} anos).")
else:
    print("Nenhuma pessoa foi cadastrada.")
```
