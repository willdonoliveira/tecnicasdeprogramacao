```python
#ex011
#Elabore um algoritmo que leia um número de entrada que indicará a quantidade de números a serem lidos. Em seguida, leia n números (conforme o valor informado anteriormente) e imprima a soma e a média aritmética dos números informados
num = int(input('Digite a quantidade de números a serem lidos: '))
soma = 0
print(f'Digite {num} números:')

for i in range(num):
    valor = float(input(f'Digite o número {i+1}: '))
    soma += valor

media = soma / num

print(f'A soma dos números informados é: {soma}')
print(f'A média dos números informados é: {media:.2f}')
```

```python
#ex013
#Elabore um algoritmo que leia um número de entrada que indicará a quantidade de registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados

num = int(input('Digite a quantidade de registros: '))
soma_m = 0
soma_f = 0
contador_m = 0
contador_f = 0

for i in range(num):
    sexo = input(f'Digite o sexo da pessoa {i+1} (M/F): ').upper()
    idade = int(input(f'Digite a idade da pessoa {i+1}: '))

    if sexo == 'M':
        soma_m += idade
        contador_m += 1
    elif sexo == 'F':
        soma_f += idade
        contador_f += 1
    else:
        print('Sexo inválido. Por favor, digite M ou F.')

media_m = 0
media_f = 0

if contador_m > 0:
    media_m = soma_m / contador_m

if contador_f > 0:
    media_f = soma_f / contador_f

print(f'A média de idade dos homens é: {media_m:.2f}')
print(f'A média de idade das mulheres é: {media_f:.2f}')
```

```python
#ex015
#Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente

numeros = []

while True:
    try:
        entrada = float(input('Digite um número real (ou 0 para encerrar): '))
        if entrada == 0:
            break
        numeros.append(entrada)
    except ValueError:
        print('Entrada inválida. Por favor, digite um número real.')

if not numeros:
    print('Nenhum número foi informado.')
else:
    numeros.sort()
    print(f'Os números informados em ordem crescente são: {numeros}')
```

```python
#ex016
#Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
#sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
#final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
#18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.

idades = []
while True:
    idade_str = input('Digite a idade do aluno (ou digite "fim" para encerrar): ')
    if idade_str.lower() == 'fim':
        break
    try:
        idade = int(idade_str)
        idades.append(idade)
    except ValueError:
        print('Entrada .+666666666666666666inválida. Por favor, digite um número inteiro para a idade ou "fim" para encerrar.')

if not idades:
    print('Nenhuma idade foi informada.')
else:
    # Idade do mais novo e mais velho
    idade_mais_novo = min(idades)
    idade_mais_velho = max(idades)

    # Contagem de alunos por idade
    alunos_mais_18 = sum(1 for idade in idades if idade > 18)
    alunos_ate_18 = sum(1 for idade in idades if idade <= 18)

    # Média aritmética
    media_aritmetica = sum(idades) / len(idades)

    # Mediana
    idades_ordenadas = sorted(idades)
    n = len(idades_ordenadas)
    if n % 2 == 0:
        mediana = (idades_ordenadas[n // 2 - 1] + idades_ordenadas[n // 2]) / 2
    else:
        mediana = idades_ordenadas[n // 2]

    print(f'Idade do aluno mais novo: {idade_mais_novo}')
    print(f'Idade do aluno mais velho: {idade_mais_velho}')
    print(f'Alunos com mais de 18 anos: {alunos_mais_18}')
    print(f'Alunos com até 18 anos: {alunos_ate_18}')
    print(f'Média aritmética das idades: {media_aritmetica:.2f}')
    print(f'Mediana das idades: {mediana:.2f}')
```
